## Task 2: Prediction Using Unsupervised Learning

In this task K-means clustering is done on the 'Iris' dataset. The dataset is clustered into 3 main categories. 

## Step 1 - Importing Libraries  

```python
from sklearn import datasets
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans 
import matplotlib.patches as mpatches
import sklearn.metrics as sm 
from mpl_toolkits.mplot3d import Axes3D 
from scipy.cluster.hierarchy import linkage,dendrogram 
from sklearn.cluster import DBSCAN 
from sklearn.decomposition import PCA 
```

## Step 2 - Loading the Dataset 

```python
iris = datasets.load_iris()
print(iris.data)
```

```python
print(iris.target_names)
```

```python
print(iris.target)
```

```python
x=iris.data
y=iris.target
```

## Step 3 - Visualising the input data and its Hierarchy

```python
#3D-plot

fig = plt.figure(figsize=(7,5))
# fig.set_figwidth(7)
# fig.set_figheight(5)
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev =48, azim =134)
ax.scatter(x[:, 3], x[:, 0], x[:, 2], edgecolor="k", s = 50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("Iris Clustering K Means=3", fontsize=15)
plt.show()

#Hierarchy clustering 

hier = linkage (x, "ward")
max_d = 7.08
plt.figure(figsize=(15,8))
plt.title("Iris Hierarchical Clustering Dendrogram", fontsize=15)
plt.xlabel("Species")
plt.ylabel("Distance")
dendrogram(
    hier,
    truncate_mode="lastp",
    p=50,
    leaf_rotation=90.,
    leaf_font_size=8.,
)

plt.axhline(y=max_d, c="k")
plt.show()



```

## Step 4- Data Preprocessing 

```python
x = pd.DataFrame(iris.data, columns=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"])
y = pd.DataFrame(iris.target, columns = ["Target"])
```

```python
x.head()
```

```python
y.head()
```

## Step 5 - Model Training 

```python
iris_k_mean_model = KMeans(n_clusters=3)
iris_k_mean_model.fit(x)
```

```python
print(iris_k_mean_model.labels_)
```

```python
print(iris_k_mean_model.cluster_centers_)
```

## Step 6 - Visualizing the Model Cluster  

```python
plt.figure(figsize=(14,6))

colors = np.array(["red", "green", "blue"])

predictedY = np.choose(iris_k_mean_model.labels_, [1, 0, 2]).astype(np.int64)

red_patch = mpatches.Patch(color="red", label="Setosa")
green_patch = mpatches.Patch(color="green", label="Versicolor")
blue_patch = mpatches.Patch(color="blue", label="Virginica")

plt.subplot(1, 2, 1)
plt.scatter(x["Petal Length"], x["Petal Width"], c=colors[y['Target']])
plt.title("Before Classification")
plt.legend(handles=[red_patch, green_patch, blue_patch])


plt.subplot(1, 2, 2)
plt.scatter(x["Petal Length"], x["Petal Width"], c=colors[predictedY])
plt.title("Model's Classfication")
plt.legend(handles=[red_patch, green_patch, blue_patch])


```

## Step 7 - Calculating the Accuracy and Confusion Matrix  

```python
sm.accuracy_score(predictedY, y['Target'])
```

```python
sm.confusion_matrix(predictedY, y['Target'])
```

### Conclusion

##### K Means Clustering has successfully clustered the dataset into distinct clusters with a good accuracy
